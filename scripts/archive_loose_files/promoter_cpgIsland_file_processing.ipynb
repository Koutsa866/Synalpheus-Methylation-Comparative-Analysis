```python
promoter_bed_fixed = "/content/drive/MyDrive/2025_SnrProj_Dir/results/gene_promoters_fixed.bed"

with open(promoter_bed, "r") as infile, open(promoter_bed_fixed, "w") as outfile:
    for line in infile:
        fields = line.strip().split("\t")
        chrom, start, end, *rest = fields
        start = int(start)
        end = int(end)

        # If it's a 0-length region, expand ¬±500 bp
        if start == end:
            start = max(0, start - 500)
            end = end + 500

        outfile.write("\t".join([chrom, str(start), str(end)] + rest) + "\n")

print("Fixed promoter BED written to:", promoter_bed_fixed)
```

```python
import os
import subprocess
from tqdm import tqdm

# Step 1: Define file paths
promoter_bed = "/content/drive/MyDrive/2025_SnrProj_Dir/results/gene_promoters_fixed.bed"
cpg_bed = "/content/drive/MyDrive/2025_SnrProj_Dir/results/cpg_methylation_overlap_trimmed_sorted.bed"
output_file = "/content/drive/MyDrive/2025_SnrProj_Dir/results/promoter_to_nearest_cpg.bed"
chunk_dir = "/content/chunks_closest"
os.makedirs(chunk_dir, exist_ok=True)

# Step 2: Read and chunk promoter BED
with open(promoter_bed, "r") as infile:
    lines = infile.readlines()

chunk_size = 10000
chunk_files = []
for i in range(0, len(lines), chunk_size):
    chunk_path = os.path.join(chunk_dir, f"promoter_chunk_{i//chunk_size}.bed")
    with open(chunk_path, "w") as chunk:
        chunk.writelines(lines[i:i+chunk_size])
    chunk_files.append(chunk_path)

# Step 3: Intersect each chunk using bedtools closest
with open(output_file, "w") as final_out:
    for chunk_path in tqdm(chunk_files, desc="üîç Finding closest CpG"):
        result = subprocess.run([
            "bedtools", "closest",
            "-a", chunk_path,
            "-b", cpg_bed,
            "-D", "a",  # reports distance from promoter (A) to CpG (B)
            "-s"        # strand-aware
        ], capture_output=True, text=True)

        if result.returncode == 0:
            final_out.write(result.stdout)
        else:
            print(f"‚ùå Error in {chunk_path}:\nSTDERR: {result.stderr}\nSTDOUT: {result.stdout}")
```

```python
import pandas as pd

# Load data
cols = list(range(6 + 6)) + ['distance']  # Adjust if needed
df = pd.read_csv(output_file, sep="\t", header=None, names=cols)

# Filter by absolute distance ‚â§ 1000 bp
df_filtered = df[df['distance'].abs() <= 1000]

# Save
filtered_output = output_file.replace(".bed", "_within1kb.bed")
df_filtered.to_csv(filtered_output, sep="\t", index=False, header=False)

print(f" Filtered results saved to: {filtered_output}")
```

```python
##  Parsing Script for promoter_to_nearest_cpg_within1kb.bed
import pandas as pd

# Path to input BED file (from bedtools closest)
input_bed = "/content/drive/MyDrive/2025_SnrProj_Dir/results/promoter_to_nearest_cpg_within1kb.bed"
output_csv = "/content/drive/MyDrive/2025_SnrProj_Dir/results/promoter_cpg_parsed.csv"

# Define column names based on BED structure
columns = [
    "promoter_contig", "promoter_start", "promoter_end", "gene_id", "dot", "strand",
    "cpg_contig", "cpg_start", "cpg_end", "methylation_status", "methylation_level", "cpg_strand",
    "distance_to_cpg"
]

# Read in the BED file (no header)
df = pd.read_csv(input_bed, sep="\t", header=None, names=columns)

# Optional: drop unnecessary columns
df = df.drop(columns=["dot"])

# Save to CSV
df.to_csv(output_csv, index=False)

print(f"‚úÖ Parsed data saved to: {output_csv}")
print(df.head())
```

```python
import pandas as pd

# Load the parsed data
input_csv = "/content/drive/MyDrive/2025_SnrProj_Dir/results/promoter_cpg_parsed.csv"
output_csv = "/content/drive/MyDrive/2025_SnrProj_Dir/results/filtered_promoter_cpg_pairs.csv"

# Read the CSV into a DataFrame
df = pd.read_csv(input_csv)

# Convert methylation_level to numeric (in case it's string type)
df["methylation_level"] = pd.to_numeric(df["methylation_level"], errors="coerce")

# Filtering conditions
filtered_df = df[
    (df["methylation_level"] >= 2) &
    (df["distance_to_cpg"] <= 1000)
]

# Save the filtered results
filtered_df.to_csv(output_csv, index=False)
print(f" Filtered results saved to: {output_csv}")
print(f" Rows retained: {len(filtered_df)} / {len(df)}")
```
